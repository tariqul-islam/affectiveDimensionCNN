```python
import sys
sys.path.insert(0, '/media/rb/Omega/0. Thesis Linux/TDD Code/caffe-action_recog/python') #To Look For .py Files in this directory
import caffe
import scipy.io
import scipy as sp
import numpy as np
from pylab import *
from sklearn.metrics import mean_squared_error #For RMSE Calculation
from math import sqrt                          #For RMSE Calculation
import pickle                                  #For saving predicted data externally
import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline
```

```python
#Model 2k, Test Video: 45, RMSE: 0.189992363378
model = 'deploy_new.prototxt';
weights = 'outputModel_iter_2000.caffemodel'; ######

net = caffe.Net(model,weights,caffe.TEST);

with open( 'List_21.txt', 'r' ) as T :
    path_list = T.readlines()

predicted_value2k = np.zeros((len(path_list), 1),dtype="f4")          ######
actual_value = np.zeros((len(path_list), 1),dtype="f4")

for i,l in enumerate(path_list):
    sp = l.split(' ')    
    f = scipy.io.loadmat(sp[0]) #sp[0] is matfilename
    X_samples = f.get('samples')
    X_samples = np.transpose(X_samples,(1,0))
    X_samples = X_samples[np.newaxis, :, :] #(n_samples, n_channels, height, width)
    net.blobs['data'].data[...] = X_samples
    out = net.forward()
    predicted_value2k[i] = net.blobs['fc2'].data            ######
    actual_value[i] = float(sp[1])
    
#_, ax1 = subplots() #Underscore is a general purpose throwaway variable
#ax2 = ax1.twinx()
#ax1.plot(arange(len(path_list)), predicted_value6k) #arange is like linspace of matlab ######
#ax2.plot(arange(len(path_list)), actual_value, 'r')
#ax1.set_xlabel('Image Number')
#ax1.set_ylabel('Prediction_Blue')
#ax2.set_ylabel('Actual_Red')

plt.plot(arange(len(predicted_value2k)), predicted_value2k, 'b',arange(len(actual_value)), actual_value, 'r')

rmse2k = sqrt(mean_squared_error(actual_value, predicted_value2k)) ########
print rmse2k

# Saving Predicted Value
with open('actual_value.pickle', 'w') as f:
    pickle.dump(actual_value, f)

# Saving Predicted Value
with open('predicted_value2k.pickle', 'w') as f:
    pickle.dump(predicted_value2k, f)

print 'Complete'
```
